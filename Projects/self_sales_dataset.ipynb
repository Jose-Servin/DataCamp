# Step 6: Grouped Summary Statistics

## Introduction to the groupby method

```python
sales_data.head()
```

# Question 3: What region has the most sales?

```python
region_sales = sales_data.groupby('Region')['Sales'].sum()
region_sales.sort_values(ascending=False)
```

# Question 4: What product category had the most sales and which one had the highest average sale? 

```python
category_sales = sales_data.groupby('Category')['Sales'].sum()
category_sales.sort_values(ascending=False)
```

```python
category_avg_sales = sales_data.groupby('Category')['Sales'].mean()
category_avg_sales.sort_values(ascending=False)
```

# Question 5: What category and sub-category had the most sales ? 

```python
category_sub_cat_sales_avg = sales_data.groupby(['Category','Sub-Category'])['Sales'].mean()
category_sub_cat_sales_avg.sort_values(ascending=False)
```

```python
# using the agg() method to see both average sale and also total sales per category/sub-category combo
cat_sub_cat_sales = sales_data.groupby(['Category','Sub-Category'])['Sales'].agg(['sum','mean'])
cat_sub_cat_sales.sort_values(['sum','mean'], ascending=[False, False])
```

# Question 6: Find average sale and average tax per category/sub-category combo

```python
cat_sub_cat_avgs = sales_data.groupby(['Category','Sub-Category'])[['Sales','taxes']].mean()
```

```python
cat_sub_cat_avgs.sort_values('Sales', ascending=False)
```

# Step 7: Pivot Tables (another way of calculating grouped summary statistics)

```python
sales_data.groupby('Category')['Sales'].mean()
```

```python
sales_data.pivot_table(values='Sales',index='Category') # by default, pivot calculates the mean of the values argument(s) 
# values is the column you want to summarize (numeric column) [] from groupby 
# index is the column you want to group by (categorical column) () from groupby 
```

## aggfun argument 

```python
# use the aggfun argument in pivot_tables to explicitly define summary statistic
sales_data.pivot_table(values='Sales', index='Category', aggfunc=[np.sum, np.mean, np.median])
```

## Pivot on two variables 

```python
sales_data.groupby(['Category','Sub-Category'])['Sales'].mean()
```

```python
sales_data.pivot_table(values='Sales', index='Category', columns='Sub-Category', aggfunc=np.mean, fill_value=0,  margins=True)
```

```python
# check the ALL column by verifying the mean sales of each category is equal to value shown
sales_data.pivot_table(values='Sales', index='Category')
```

## Multi-Level indexing (hierarchical indexes)

### Setting Category and Sub-Category as indexes 

```python
category_sub_category_index = sales_data.set_index(['Category','Sub-Category'])
category_sub_category_index.head()
```

### subsetting outler level indexes (Category)

```python
# Pull data for all Furniture and Office Supplies product category 
furniture_and_office = category_sub_category_index.loc[['Furniture','Office Supplies']]
furniture_and_office
```

### subsetting inner level indexes

```python
# pull data for Furniture/Tables and Office Supplies/Paper (passing a list of tuple pairs)
inner_subsetting = category_sub_category_index.loc[[('Furniture','Tables'),('Office Supplies','Paper')]]
inner_subsetting
```

## Sorting by index 

```python
# sorting category DataFrame in ABC order (outter to inner index levels in ascending order)
category_sub_category_index.sort_index()
```

```python
# sort Sub-Category in ascdending order and Category in descending order 
category_sub_category_index.sort_index(level=['Sub-Category','Category'], ascending=[True,False])
```
