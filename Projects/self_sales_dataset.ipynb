# Step 2: Sorting and Subsetting DataFrames

## Using .sort_values()

```python
# sorting by ascending values (smalles to largest)
sales_data.sort_values('Sales')
```

```python
# Sorting by sales larges to smallest
sales_data.sort_values('Sales', ascending =False)
```

```python
# Sorting by categorical columns
sales_data.sort_values(['Category','Sub-Category'], ascending=[True, False])
```

## Subsetting DataFrames (pull needed info)

### column subsetting

```python
order_region_df = sales_data[['Order ID', 'Region']]
order_region_df
```

### row subsetting

```python
west_orders = sales_data[sales_data['Region']== 'West']
west_orders
```

```python
orders_over_10k = sales_data[sales_data['Sales'] >= 10000]
orders_over_10k
```

```python
order_over_10k_west = sales_data[(sales_data['Sales'] >= 10000) & (sales_data['Region'] == 'West')]
order_over_10k_west
```

### categorical subsetting

```python
# First start by looking at unique categorical values 
sales_data.Category.unique()
```

```python
# pull Technology and Furniture data
tech_and_futniture = sales_data[sales_data['Category'].isin(['Technology','Furniture'])]
tech_and_futniture.sort_values('Sales', ascending=False)
```
