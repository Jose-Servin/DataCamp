# Matrix Plots

* most common type is the heat map
* can be use to quickly see trends in the data 

## heatmap() 

* requires data to be in a grid format 
* pandas crosstab() is used to manipulate data 

```python
df.head(3)
```

```python
pd.crosstab(
    df['model_year'],
    df['origin'],
    values=df['mpg'],
    aggfunc='mean'
).round(0)
```

```python
sns.heatmap(
    pd.crosstab(
    df['model_year'],
    df['origin'],
    values=df['mpg'],
    aggfunc='mean'
).round(0)
)
```

## Customize a heatmap

```python
df_crosstab = pd.crosstab(
    df['model_year'],
    df['origin'],
    values=df['mpg'],
    aggfunc='mean'
).round(0)
```

```python
sns.heatmap(
    df_crosstab,
    annot=True,  # shows values of individual cells 
    #fmt='d', # ensures results are displayed as integers 
    cmap='YlGnBu', # custom color map "cmap"
    cbar = False, # color bar 
    linewidths= 0.5 # places small spacing between the cells 
)
```

## Centering a Heatmap

* shifting the overall color scheme of the heatmap 

```python
sns.heatmap(
    df_crosstab,
    annot=True,  # shows values of individual cells 
    #fmt='d', # ensures results are displayed as integers 
    cmap='YlGnBu', # custom color map "cmap"
    cbar = False, # color bar 
    linewidths= 0.5, # places small spacing between the cells 
    center=df_crosstab.loc[82,'europe']
)
```

# Using FacetGrid, factorplot and Lmplot 

## Facetgrid

* The foundation of many data aware grids
* Allows the user to control how data is distributed across columns, rows and hue
* once a Facetgrid is created, the plot type must be mapped to the grid

### Facetgrid categorical example

```python
g = sns.FacetGrid(data=df, col='origin')
g.map(
    sns.boxplot,
    'horsepower',
    order = ['usa','japan','europe']
    )
```

## Facetgrid for regression and scatter plots

```python
g = sns.FacetGrid(data=df, col='origin') # col is the categorical variable 
g.map(plt.scatter, 'horsepower', 'displacement') # two numeric variables being compared on a scatter plot 
```

```python
g = sns.FacetGrid(data=df, col='origin') # col is the categorical variable 
g.map(sns.regplot, 'horsepower', 'displacement') # two numeric variables being compared on a scatter plot 
```

## lmplot (one step option)

* plots scatter and regression plots on a FacetGrid

```python
sns.lmplot(
    data=df,
    col='origin',
    x='horsepower',
    y='displacement',
    fit_reg=False # disables regression line
)
```

## Factorplot (one step option)

* a simpler way to use a FacetGrid for categorical data
* combines Facetting and mapping process into 1 function

```python
sns.factorplot(
    data = df,
    x='horsepower',
    col='origin',
    kind='box'
)
```

## FacetGrid Examples

```python
g = sns.FacetGrid(
    data=df, 
    col='origin',
    col_order = ['usa','europe','japan']
)

g.map(
    sns.pointplot,
    'horsepower'
)
```

## FacetorPlot Example

* The factorplot is often more convenient than using a FacetGrid for creating data aware grids.

```python
sns.factorplot(
    data=df,
    x='horsepower',
    col='origin',
    kind= 'box',
    col_order = ['usa','europe','japan']
)
```

## FacetGrid scatter Example

```python
g = sns.FacetGrid(
    data=df,
    col = 'origin',
    col_order= ['usa','europe','japan']
)

g.map(
    plt.scatter,
    'horsepower',
    'displacement'
)
```

# Using PairGrid and pairplot

## PairGrid

* shows pairwise relationships between data elements
* follows similar API to FacetGrid

```python
g = sns.PairGrid(
    data=df,
    vars=['horsepower','displacement']
)
g.map(plt.scatter)
```

## Customizing the PairGrid diagonols

```python
g = sns.PairGrid(
    data=df,
    vars=['horsepower','displacement']
)
g.map_diag(plt.hist) # main diagonol 
g.map_offdiag(plt.scatter) # other diagonol
```

## Pairplot

* shortcut for creating PairGrid

```python
sns.pairplot(
    data=df,
    vars = ['horsepower', 'displacement'],
    kind = 'scatter', # main diagonol
    diag_kind='hist' # other diagonol 
)
```

### customizing a pairplot

* pairplot with regression 

```python
sns.pairplot(
    data=df,
    vars = ['horsepower', 'displacement'],
    kind = 'reg', # main diagonol
    diag_kind='hist', # other diagonol 
    hue = 'model_year'
)
```

# Using JointGrid and jointplot

## JointGrid

```python
g = sns.JointGrid(
    data=df,
    x='horsepower',
    y='displacement'
)

g.plot(
    sns.regplot,
    sns.distplot 
    # sns.histplot
)
```

## Advanced JointGrid

```python
g = sns.JointGrid(
    data=df,
    x='horsepower',
    y='displacement'
)

g = g.plot_joint(sns.kdeplot)
g = g.plot_marginals(sns.kdeplot, shade =True)
# g= g.annotate(pearsonr)
```

## Jointplot

```python
sns.jointplot(
    data=df,
    x='horsepower',
    y='displacement',
    kind='hex'
)
```

# Selecting Seaborn Plots

* Univariate Distribution Analysis 
    * distplot() is the best place to start analysis of one variable 
    * rugplot() and kdeplot() can be useful alternatives (components of distplot)


* Regression Analysis : relationship between two variables
    * lmplot() performs regression analysis and supports Facetting
    * regplot and scatter plot are lmplot components
    * best option to see linear relationships between data 

* Categorical Plot
    * factorplot represents the best option compared to FacetGrid
    

* Regression Analysis Plots Part 2
    * pairplot and jointplot offer useful ways to see how two variables interact with eachother
