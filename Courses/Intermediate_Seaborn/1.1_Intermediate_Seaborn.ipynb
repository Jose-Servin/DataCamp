# Using the Distribution Plot

## Creating a standard Histogram

```python
sns.distplot(df['horsepower'], kde=False, bins=10)
```

## Alternative data distributions

* A rug plot is an alternative way to view the distribution of data

```python
sns.distplot(df['horsepower'], kde=False, rug=True)
```

```python
sns.distplot(df['horsepower'], kde=True, kde_kws={'shade':True})
```

# Regression Plots in Seaborn

* The regplot function generates a scatter plot with a regression line.
* The data, x and y variables must be defined. 

```python
sns.regplot(data=df, x='displacement', y='horsepower')
```

## sns.lmplot() represents a higher level plot for the regplot()

* Faceting is the use of plotting multiple graphs while changing a single variable

```python
sns.lmplot(data=df, x='displacement', y='horsepower', col='origin')
```

```python
sns.lmplot(data=df, x='displacement', y='horsepower', hue='origin')
```

# Colors in Seaborn

* Seaborn supports Matplotlib color codes by passing color_codes=True to sns.set()

```python
sns.set(color_codes=True)
sns.distplot(df['horsepower'], color='g')
```

## Color Pallets

```python
for p in sns.palettes.SEABORN_PALETTES:
    print(p)
```

## Displaying Palettes

* sns_palplot() function displays the palette
* sns.color_palette() returns the current palette

```python
for p in sns.palettes.SEABORN_PALETTES:
    sns.set_palette(p)
    sns.palplot(sns.color_palette())
    plt.show()
```

## Defining Custom Palettes

### Circular Palette: when data is not ordered

```python
sns.palplot(
    sns.color_palette('Paired', 12) # 12 is the argument for number of colors returned
)
```

### Sequential Palette: when the data has a consistant range from high to low 

```python
sns.palplot(
    sns.color_palette('Blues',12)
)
```

### Diverging colors: when both the low and high values are interesting 

```python
sns.palplot(
    sns.color_palette('BrBG', 12)
)
```

# Customizing with matplotlib

## matplotlib axes

* Axes can be passed to seaborn functions

```python
fig, ax = plt.subplots()
sns.histplot(df['horsepower'], ax =ax) # matplotlib function passed to seaborn histplot() function 
ax.set(xlabel = 'Horsepower Distribution') # matplotlib function to set xlabel 
```

## Further Customization

```python
fig, ax = plt.subplots()
sns.histplot(df['horsepower'], ax =ax) # matplotlib function passed to seaborn histplot() function 
ax.set(
    xlabel = 'Horsepower Distribution',
    ylabel = 'Count',
    xlim=(50, 175),
    title='Histogram of Horsepower'
    ) # matplotlib function to set xlabel 
```

# Combining Plots

```python
df.origin.unique()
```

```python
fig, (ax0, ax1) = plt.subplots(
    nrows=1, ncols=2, sharey=True, figsize = (7,4)
)

sns.histplot(df.query('origin == "usa"')['horsepower'], ax=ax0)
sns.histplot(df.query('origin == "japan"')['horsepower'], ax=ax1)

ax1.set(
    xlabel = 'Japan',
    title = 'Horsepower Histogram Japan'
)
ax1.axvline(x=80, label='Japan vline')
ax0.set(
    xlabel = 'USA',
    title = 'Horsepower Histogram USA'
)
ax0.axvline(x=110, label='USA vline')
ax1.legend()
ax0.legend()
```
