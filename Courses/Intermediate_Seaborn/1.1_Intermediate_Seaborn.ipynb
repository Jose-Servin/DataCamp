# Matrix Plots

* most common type is the heat map
* can be use to quickly see trends in the data 

## heatmap() 

* requires data to be in a grid format 
* pandas crosstab() is used to manipulate data 

```python
df.head(3)
```

```python
pd.crosstab(
    df['model_year'],
    df['origin'],
    values=df['mpg'],
    aggfunc='mean'
).round(0)
```

```python
sns.heatmap(
    pd.crosstab(
    df['model_year'],
    df['origin'],
    values=df['mpg'],
    aggfunc='mean'
).round(0)
)
```

## Customize a heatmap

```python
df_crosstab = pd.crosstab(
    df['model_year'],
    df['origin'],
    values=df['mpg'],
    aggfunc='mean'
).round(0)
```

```python
sns.heatmap(
    df_crosstab,
    annot=True,  # shows values of individual cells 
    #fmt='d', # ensures results are displayed as integers 
    cmap='YlGnBu', # custom color map "cmap"
    cbar = False, # color bar 
    linewidths= 0.5 # places small spacing between the cells 
)
```

## Centering a Heatmap

* shifting the overall color scheme of the heatmap 

```python
sns.heatmap(
    df_crosstab,
    annot=True,  # shows values of individual cells 
    #fmt='d', # ensures results are displayed as integers 
    cmap='YlGnBu', # custom color map "cmap"
    cbar = False, # color bar 
    linewidths= 0.5, # places small spacing between the cells 
    center=df_crosstab.loc[82,'europe']
)
```
